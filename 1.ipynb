```python
import csv

attributes=[['Sunny','Rainy'],['Warm','Cold'],['Normal','High'],['Strong','Weak'],['Warm','Cool'],['Same','Change']]

num_attributes=len(attributes)
print("\n the most general hypothesis:('?','?','?','?','?','?')\n")
print("\n the most specific hypothesis:('0','0','0','0','0','0')\n")
a=[]
yes=0

print("\n the given trainig data set\n")
reader=csv.reader(open('1.csv'))
for row in reader:
    a.append(row)
    print(row)
         
print("\n the initial value of hypothesis:")
hypothesis=['0']*num_attributes
print(hypothesis)

for i in range(len(a)):
    if a[i][-1]=='No':
        pass
    if a[i][-1]=='Yes':
        yes=1
        for j in range(num_attributes):
            hypothesis[j]=a[i][j]
    if(yes==1):
        break

print("\n the first positive sample of hypothesis is:\n")
print(hypothesis)
print("\nFind S:Finding a maximally specific hypothesis\n")

for i in range(len(a)):
    
    if a[i][-1]=='Yes':
        
        for j in range(num_attributes):
        
            if a[i][j]!= hypothesis[j]:
                
                hypothesis[j]='?'
            else:
                
                hypothesis[j]=a[i][j]
        print("for training example no:{0} the hypothesis".format(i),hypothesis)
print("\n maximally specific hypothesis for a given training example:\n")  
print(hypothesis)
```
