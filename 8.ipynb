```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
```

```python
iris = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris['data'], iris['target'], random_state=20)
```

```python
kmeans = KMeans(n_clusters=3, random_state=20)
kmeans.fit(X_train)
km_preds = kmeans.predict(X_test)
print('KMeans score:\n', classification_report(y_test, km_preds))

plt.subplot(1, 2, 1)
plt.title('KMeans')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.scatter(X_test[km_preds == 0, 0], X_test[km_preds == 0, 1], label='setosa', cmap='gist-rainbow')
plt.scatter(X_test[km_preds == 1, 0], X_test[km_preds == 1, 1], label='versicolor', cmap='gist-rainbow')
plt.scatter(X_test[km_preds == 2, 0], X_test[km_preds == 2, 1], label='virginica', cmap='gist-rainbow')

plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], color='black', label='Cluster centers')
plt.legend()
```

```python
gm = GaussianMixture(n_components=3, random_state=20)
gm.fit(X_train)
gm_preds = gm.predict(X_test)
print('Gaussian Mixture score:\n', classification_report(y_test, gm_preds))

plt.subplot(1, 2, 2)
plt.title('Gaussian Mixture')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.scatter(X_test[gm_preds == 0, 0], X_test[gm_preds == 0, 1], label='setosa', cmap='gist-rainbow')
plt.scatter(X_test[gm_preds == 1, 0], X_test[gm_preds == 1, 1], label='versicolor', cmap='gist-rainbow')
plt.scatter(X_test[gm_preds == 2, 0], X_test[gm_preds == 2, 1], label='virginica', cmap='gist-rainbow')

plt.scatter(gm.means_[:, 0], gm.means_[:, 1], color='black', label='Cluster centers')
plt.legend()
plt.show()
```
